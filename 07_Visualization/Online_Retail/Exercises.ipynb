```python
online_rt = pd.read_csv("Online_Retail.csv", encoding='latin1')
```

### Step 4. Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK

```python
ds = online_rt.query('Country != \'United Kingdom\'')
ds_dict = ds.groupby("Country").sum()
countries = ds_dict.sort_values(by='Quantity', ascending=False)[0:10]
countries['Quantity'].plot(kind='bar')
plt.xlabel('Countries')
plt.ylabel('Quantity')
plt.title('10 Countries with much order')
plt.show()
```
