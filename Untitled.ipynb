```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.linspace(-np.pi, np.pi, 10)
C,S = np.sin(X), np.cos(X)
```

```python
plt.plot(X, C)
plt.plot(X, S)
```

```python
Y = np.linspace(-np.pi, np.pi, 264)
C1, S1 = np.sin(Y), np.cos(Y)
plt.plot(Y, C1)
plt.plot(Y, S1)
plt.show()
```

```python
#create a figure of 8x6, using 100 dots per pixel 
plt.figure(figsize=(8, 6), dpi=80)

plt.subplot(111)

X = np.linspace(-np.pi, np.pi, 264, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")

plt.plot(X, S, color="green", linewidth=2.0, linestyle="-")

plt.xlim(-4.0, 4.0)
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))

plt.ylim(-1.0, 1.0)
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

plt.show()
```

```python
plt.figure(figsize=(10, 6), dpi=100)
plt.plot(X, C, color="blue", linewidth=2.0, linestyle="-")
plt.plot(X, S, color="red", linewidth=2.0, linestyle="-")
plt.show()
```

```python
plt.figure(figsize=(10, 6), dpi=100)
plt.plot(X, C, color="blue", linewidth=2.0, linestyle="-")
plt.plot(X, S, color="red", linewidth=2.0, linestyle="-")
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])
plt.show()
```

```python
n = 12
X = np.arange(n);
Y1 = np.random.uniform(0.5, 1.0, n) * (1 - X/float(n))
Y2 = np.random.uniform(0.5, 1.0, n) * (1 + X/float(n))
plt.figure(figsize=(10,6), dpi=100)
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor="white")
plt.bar(X, -Y1, facecolor='#889900', edgecolor="white")
for x,y in zip(X,Y1):
    plt.text(x+0.025, y+0.005, '%.2f'%y, ha="center", va="bottom")
plt.ylim(-1.25, 1.25)
plt.show()
```
